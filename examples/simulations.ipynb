```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sampling_techniques as st
import seaborn as sns
```

```python
def poids_alea(taille):

    tmp = np.random.choice(np.arange(0, 7, 1))

    if np.random.choice([0, 1]) == 0:

        return (taille - 100 - ((taille  - 150)/4)) + tmp

    else:

        return (taille - 100 - ((taille  - 150)/4)) - tmp
```

# Données simulées

```python
tab_tech = ["cluster", "custom", "k_mean", "kennard_stone", "random", "SPlit", "spxy", "systematic"]

np.random.seed(0)

taille = np.random.randn(200) * 11 + 163

f_test = pd.DataFrame(data=taille, columns=["Taille (cm)"])

l_test = pd.DataFrame(data=[poids_alea(t) for t in taille], columns=["Poids (kg)"])
```

# Illustration techniques sampling

```python
plt.figure(figsize=(20, 20))

i = 1

for tech in tab_tech:

    f_tr, f_ts, l_tr, l_ts = st.sampling_train_test_split(f_test, l_test, test_size=0.1, tech=tech)

    plt.subplot(4, 2, i)
    plt.title("Splitting d'un jeu de données simulé par la technique : {}".format(tech), pad=0)
    plt.xlabel("Taille (cm)")
    plt.ylabel("Poids (kg)")
    plt.scatter(x=f_tr, y=l_tr, c='lightblue', edgecolors='black')
    plt.scatter(x=f_ts, y=l_ts, c='red', edgecolors='black', )
    #plt.savefig("Splitting_{}.png".format(tech), bbox_inches='tight')
    i+=1

#plt.title("Scatterplot d'un jeu de données simulé", pad=20)
#plt.scatter(x=f_test, y=l_test, c='lightblue', edgecolors='black', )
#plt.savefig("data_simulation.png".format(tech), bbox_inches='tight')

plt.show()
```
